```python
#Step-1 Importing the Dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Step-2 Data Collection and Processing

#loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('heart.csv')
```

```python
#print first 5 rows of the dataset
heart_data.head()
```

```python
#print last 5 rows of the dataset
heart_data.tail()
```

```python
#number of rows and columns in the dataset
heart_data.shape
```

```python
#getting some info about the data
heart_data.info()
```

```python
#checking if there is any missing values
heart_data.isnull().sum()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
# checking the distribution of Target variable
# 1-having heart disease(defective heart)
# 0-not having heart disease
heart_data['target'].value_counts()
```

```python
#Splitting the features and targets
X = heart_data.drop(columns='target', axis=1)
Y = heart_data['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Step-4 Splitting the Training data and Test data

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Step-5 To train our Machine Learning Model
#We are using Logistic Regression Model because it is used for binary classification

model = LogisticRegression()
```

```python
# training the LogisticRegression model with Training data
model.fit(X_train, Y_train)
```

```python
#Model Evaluation
#Accuracy Score


#First check accuracy on traing data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
#Accuracy of more than 75% is considered to be good
print("Accuracy on Training data:: ",training_data_accuracy)
```

```python
#Now check accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
#Accuracy of more than 75% is considered to be good
print("Accuracy on Test data:: ",test_data_accuracy)
```

```python
#Accuracy score of training data and test data must be have less difference otherwise our model is said to overfitted
#Overfitted - When the difference between accuracy score of training and test data is too large
```

```python
#Step-5 Building a Predictive System

input_data = (53,1,0,140,203,1,0,155,1,3.1,0,0,3)

#Change the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

#Reshape the numpy array as we are predicting for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)

print(prediction)

if(prediction[0] == 0):
    print("The Person is fit and does not have heart disease")
else:
    print("The Person is having heart disease")
```
